```python
img.shape
def do_convolution(img):
    img_new=np.zeros(img.shape[:2])
    for n in range(img.shape[0]-1):
        for m in range(img.shape[1]-1):
            sum=0
            for k in range(-1,1):
                for l in range(-1,1):
                    sum+=img[n-k,m-l]
            img_new[n,m]=int(sum*1.0/9)
    return img_new
```

```python
img_new=do_convolution(img)
```

```python
img_new
```

```python
plt.imshow(img_new, cmap=cm.gray)
plt.title('convolutional')
plt.show()
```
